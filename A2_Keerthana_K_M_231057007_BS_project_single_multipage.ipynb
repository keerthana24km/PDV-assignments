```python
#Extract question details from a Single Web Page

URL = "https://www.rocketstem.org/2020/04/28/quizme-hubble-space-telescope/"

r = requests.get(URL)

soup = BeautifulSoup(r.content, 'lxml')

attr = {'class': 'hdq_question'} #attribute provided here is class
var2 = soup.findAll('div', attrs=attr) #finding all the div tags having their class name as 'hdq_question'
#print(var2)

question_details = [] #List to hold all the questions and it's details
for row in var2:
  ques_det = {} #Dictionary to hold key-value pairs of attributes of a question and it's corresponding values
  ques_det['Question_id'] = row['id']
  ques_det['Question_type'] = row['data-type']
  ques_det['Marks'] = row['data-weight']
  question_details.append(ques_det) #Adding all the dictionary values to the list

print(question_details)

filename = 'quiz_questions_details.csv'
with open(filename, 'w', newline='') as f:
	w = csv.DictWriter(f,['Question_id','Question_type','Marks']) #Writing the values of dictionary to a .csv file under the respective headings
	w.writeheader()
	for ques_det in question_details:
		w.writerow(ques_det)

```

```python
#First Page Data Extraction

BASE_URL = "https://solarcooking.fandom.com" #base url
URL = BASE_URL+"/wiki/Recipes" #absolute url
#Here, relative url is the additionally added url, "/wiki/Recipes"

r = requests.get(URL)

soup = BeautifulSoup(r.content,'lxml')

div = soup.find('div', attrs = {'class':'mw-parser-output'}) #base class for extraction of data
ul = div.ul
li = ul.findAll('li')

types = []
for row in li:
  food_types = {}
  food_types['Title'] = row.text
  food_types['Link'] = BASE_URL + row.a['href']
  types.append(food_types)

print(types)

filename = 'recipe_links.csv'
with open(filename, 'w', newline='') as f:
	w1 = csv.DictWriter(f,['Title','Link']) #Writing the values of dictionary to a .csv file under the respective headings
	w1.writeheader()
	for row in types:
		w1.writerow(row)
```

```python
#Multiple Web Page Data Extraction

recipe_links = []
for link in li:
  recipe_links.append(BASE_URL + link.a['href']) #getting all the recipe links

#print(recipe_links)
print()

for links in recipe_links: # first for loop to scan all the links
  r2 = requests.get(links)
  soup = BeautifulSoup(r2.content,'lxml')
  span = soup.findAll('a', attrs = {'class':'external text'})
  Languages = []
  for row in span: # second for loop to extract the types of languages
    lang = {}
    lang['Name'] = row.text
    lang['Link'] = row['href']
    Languages.append(lang)
  #print(Languages)
  #print()

filename = 'recipe_languages.csv'
with open(filename, 'w', newline='') as f:
	w2 = csv.DictWriter(f,['Name','Link']) #Writing the values of dictionary to a .csv file under the respective headings
	w2.writeheader()
	for row in Languages:
		w2.writerow(row)
```
